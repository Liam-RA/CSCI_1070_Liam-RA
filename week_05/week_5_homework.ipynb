```python
#I am aware that this code is unbelievably useless to any other Dataframe, a better way to do this would be to improt a list of
# all US states then compare that list to each response, same for other countries instead of just 'international'.
#This would take a lot of time to proccess each request though.

#1


#f['location'] = np.where('Missouri' in df['Place_of_origin'] )
def origin_sort(DataFrame):
    _ = DataFrame['Place_of_origin'].tolist()
    
    for i in _:
        if 'Missouri' in i or 'Illinois' in i or 'New Mexico' in i or 'Minnesoda'in i or 'United States' in i:
            DataFrame['location'] = 'United States'
            _[_.index(i)] = 'United_States'
        else:
            _[_.index(i)] = 'International'

    DataFrame = DataFrame.assign(Place_of_origin = lambda x: _)

    return DataFrame
    
origin_sort(df_fixed)
```

```python
# Same thing as function above, Not a particularly useful function to apply to any other dataframe
# A better way of doing this would probably be
# first a function that scans the df for all numbered responses in the pets column
# then take the max value, have a loop repeat until that max value.
# the loop should check for the number of interest in the responce and replace, then increment and repeat until it reaches max value

#2

def pets_sort(DataFrame):

    _ = DataFrame['Pets'].fillna('0').tolist()
    
    for i in _:
        if ('one' in i or '1' in i) and ('two' in i or '2' in i):
            _[_.index(i)] = '3'
        elif 'zero' in i or 'no' in i or 'No' in i or '0' in i:
            _[_.index(i)] = '0'
        elif 'One' in i or 'one' in i or '1' in i:
            _[_.index(i)] = '1'
        elif 'Two' in i or 'two' in i or '2' in i:
            _[_.index(i)] = '2'
        elif 'Husky' in i: #This could be improved by comparing to a dataset of dog breeds, then printing the amount of matches
            _[_.index(i)] = '1'
        else:
            _[_.index(i)] = '0'

    DataFrame = DataFrame.assign(Pets = lambda x: _)
    
    return DataFrame

pets_sort(df_fixed)
```

```python
#3

# I know that this isn't 100% standardized, but I defenitly could stadardize the whole df using this method, I'm just not sure toh ow categorize 
# foods like 'braownies' or 'steak', I can't really find a country of origin

# To improve this code you could again compare the responses to an imported list of foods from different countries, though for this one
# There could be some issues with overlapping categories, as well, some foods dont have origins in one particular country

def food_sort(DataFrame):

    _1 = list(DataFrame['Favorite_food'])

    _ = [x.lower() for x in _1]
    
    for i in _:
        if 'pasta' in i or 'italian' in i or 'pizza' in i:
            _[_.index(i)] = 'Italian'
        elif 'mexican' in i:
            _[_.index(i)] = 'Mexican'
        elif 'hot wings' in i or 'burger' in i:
            _[_.index(i)] = 'American'
        elif 'injera' in i:
            _[_.index(i)] = 'Etheopian'
        elif 'gambian' in i:
            _[_.index(i)] = 'Gambian'
        elif 'kabsa' in i or 'meddleastern' in i:
            _[_.index(i)] = 'Middleastern'
        elif 'vietnamese' in i:
            _[_.index(i)] = 'Vietnamese'
        else:
            _[_.index(i)] = i

    proxy_df = pd.DataFrame(_)
    
    DataFrame = DataFrame.assign(Favorite_food = lambda x: proxy_df)

    return DataFrame

food_sort(df_fixed)
```

```python
#4

def major_sort(DataFrame):

    _1 = DataFrame['Major'].tolist()

    _ = [x.lower() for x in _1]

    for i in _:
        if 'data science' in i or 'computer science' in i or 'comp sci' in i or 'computer sci' in i or 'comp science' in i or 'biochemistry' in i or 'biology' in i:
            _[_.index(i)] = 'School_of_Science_and_Engineering'
        elif 'math' in i or 'forensic' in i:
            _[_.index(i)] = 'College_of_Arts_and_Sciences'
        elif 'health' in i:
            _[_.index(i)] = 'Doisy_College_of_health_Sciences'
        elif 'nursing' in i:
            _[_.index(i)] = 'School_of_Nursing'
        elif 'buissness' in i or 'finance' in i or 'economics' in i or 'buisness':
            _[_.index(i)] = 'School_of_Buissness'
        elif 'criminology' in i:
            _[_.index(i)] = 'School_of_Social_Work'
        else:
            _[_.index(i)] = i

    DataFrame = DataFrame.assign(Major = lambda x:_)

    return DataFrame

major_sort(df_fixed)
```
