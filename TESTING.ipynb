```python
class candy:
    def __init__(self,
                 _brand,
                _is_caramel=False,
                 _has_nuts=False,
                 _calories=1300,
                 _type='Sour',
                 _size='large'):
        self.type=_type
        self.brand=_brand
        self.is_caramel = is_caramel
        self.type = _type

    def display(self):
        print(f'my favorite candy is {self.brand}{self.type}')
```

```python
favorite_candy = candy(_brand='none')
favorite_candy.display()
```

```python
@staticmethod
def calories(size,n):
    if size == 'Large':
        calories = 500
    elif size == 'regular':
        calories = 750
    elif size == 'fun':
        calories = 1250
    else:
        calories = 0
    calories=calories*n
    return calories
```
